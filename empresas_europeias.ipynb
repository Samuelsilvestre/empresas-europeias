```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv('./dataset_empresas_europeias - adicione a tabela algumas empresas nordicas respe....csv', encoding='utf-8')
df.columns
```

```python
df['Receita (€ Bilhões)'] = df['Receita (€ Bilhões)'].apply(lambda x: float(x.replace(',','.')))
```

```python
df.dtypes
```

```python
# analise da receita e setor das empresas europeias
Analise_setor = df.groupby(by=['Setor']).agg(sum=('Receita (€ Bilhões)','sum'))
Analise_setor
```
